# Ensembling

A lot of model combinations have been tested. Below, is just the one which generated the best submission

```python
#Best submission
first_clipped = pd.read_csv("first_clipped_0.099_47.csv")
second_clipped = pd.read_csv("second_clipped_0.099_47.csv")
a_min, a_max = 0.099, 47
submission['target_pct_vunerable']=np.clip((first_clipped['target_pct_vunerable']+second_clipped['target_pct_vunerable'])/2, a_min=a_min, a_max=a_max)
submission.to_csv(f'assembling_uniform_first_second_clipped_{a_min}_{a_max}.csv', index = False)
```
